```python
# Task 2: Identify the city with the highest number of restaurants
city_counts = data['City'].value_counts()  # Count the number of restaurants in each city
display(city_counts)
```

```python
# Calculate the average rating for restaurants in each city
average_ratings_by_city = data.pivot_table(values='Aggregate rating', index='City', aggfunc='mean')

# Display the top 5 cities with their average ratings
print("\n=== Average Rating for Restaurants in Each City ===")
print(average_ratings_by_city.head(5))
```

```python
# Calculate the number of restaurants per city
city_counts = data['City'].value_counts()

# Find the city with the highest number of restaurants
city_with_highest_restaurants = city_counts.idxmax()  #idxmax() is short for "index max", meaning it returns the city name where the count is the highest.
max_restaurant_count = city_counts.max() # max(): This function returns the maximum value in the series (in this case, the highest number of restaurants).

# Display the results
print("=== City with the Most Restaurants ===")
print("City:", city_with_highest_restaurants)
print("Number of Restaurants:", max_restaurant_count)
```

```python
# Find the city with the highest average rating
city_with_highest_rating = average_ratings_by_city.idxmax()
highest_average_rating = average_ratings_by_city.max()


print("\n=== City with the Highest Rating ===")
print("City:", city_with_highest_rating)
print("Average Rating:", highest_average_rating)
```

```python
# Analyze the most common combinations of cuisines
cuisine_combinations = data['Cuisines'].value_counts()

# Calculate the average rating for each cuisine combination
average_rating_by_cuisine = data.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False)

# Display the top 10 most common combinations and corresponding average ratings
top_cuisine_combinations = cuisine_combinations.head(10)
top_ratings_for_top_cuisines = average_rating_by_cuisine.loc[top_cuisine_combinations.index]

top_cuisine_combinations, top_ratings_for_top_cuisines
```

#Task - 4

```python
# Identify restaurant chains by counting occurrences of restaurant names
restaurant_chains = data['Restaurant Name'].value_counts()

# Filter for chains with more than one location
chains_with_multiple_locations = restaurant_chains[restaurant_chains > 1]

# Analyze ratings and popularity for these chains
chain_ratings_popularity = data[data['Restaurant Name'].isin(chains_with_multiple_locations.index)]\
    .groupby('Restaurant Name')\
    .agg(Average_Rating=('Aggregate rating', 'mean'), Average_Votes=('Votes', 'mean'))\
    .sort_values(by='Average_Rating', ascending=False)

chains_with_multiple_locations.head(10), chain_ratings_popularity.head(10)
```

```python
# Analyze the relationship between price range and availability of services

# Convert relevant columns to numeric or categorical
data['Price range'] = data['Price range'].astype(int)
data['Has Table booking'] = data['Has Table booking'].map({'Yes': 1, 'No': 0})
data['Has Online delivery'] = data['Has Online delivery'].map({'Yes': 1, 'No': 0})

# Calculate average service availability by price range
table_booking_by_price = data.groupby('Price range')['Has Table booking'].mean()
online_delivery_by_price = data.groupby('Price range')['Has Online delivery'].mean()

table_booking_by_price, online_delivery_by_price
```
